# Text extraction

Extract the text from an external source. Example: a Wikipedia article, text document, Word, PDF, Powerpoint, Tweet
You should produce results for at least 50 different texts.

Find the filename and extension of all text data from a source.

```python
import glob
import os
import textract
```

```python
dir = '/Users/user/Desktop/application/data/'
```

```python
files = glob.glob(dir+'*')
```

```python
counter = 1
for filename in files:
    (shortname, extension) = os.path.splitext(filename)
    text = textract.process(filename)
    print("We're printing example for {}".format(extension))
    print(text)
    counter += 1
    print("\n") 
    print(counter)
    print("\n") 
```
